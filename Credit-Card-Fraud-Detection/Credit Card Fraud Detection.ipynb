### Input Split

```python
x=data.drop(columns=['Class'],axis=1)
y=data['Class']
```

### Standard Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
x_scaler=sc.fit_transform(x)
x_scaler[-1]
```

### Model Training 

##### so we use 3 models to test the data to check which perform the best on the dataset
1. Logistics Regression
2. Random Forest Classifier
3. XGB Classifier

#### Logistic Regression

```python
# Train test split
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, f1_score
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaler,y,test_size=0.25,random_state=42,stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
# training
model.fit(x_train,y_train)
# testing
y_pred=model.predict(x_test)
```

```python
print(classification_report(y_test,y_pred))
print('F1 Score',f1_score(y_test,y_pred))
```

so in the classification report we get the accuracy 100 % due to the imbalanced data but the f1 score is 
69 % which is not actually good so we test another model on the dataset

#### 2. Random Forest Classfier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier()
```

```python
# training
model.fit(x_train,y_train)
# testing
y_pred=model.predict(x_test)
```

```python
print(classification_report(y_test,y_pred))
print('F1 Score',f1_score(y_test,y_pred))
```

##### In the Random Forest Model the f1 score is improved to 83 %

### Class Imbalancement

```python
sb.countplot(y_train)
```

```python
# balance the class with equal distribution
```

```python
!pip install imblearn
from imblearn.over_sampling import SMOTE
```

```python
over_sample=SMOTE()
x_smote,y_smote=over_sample.fit_resample(x_train,y_train)
```

```python
sb.countplot(y_smote)
```

#### 1.Logistic Regression(after class balance)

```python
# Train test split

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, f1_score

x_train,x_test,y_train,y_test=train_test_split(x_scaler,y,test_size=0.25,random_state=42,stratify=y)

from sklearn.linear_model import LogisticRegression
model=LogisticRegression()

# training
model.fit(x_smote,y_smote)
# testing
y_pred=model.predict(x_test)

print(classification_report(y_test,y_pred))
print('F1 Score',f1_score(y_test,y_pred))
```

#### 2. Random forest Classifier(after class balance)

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier(n_jobs=1)
```

```python
# training
model.fit(x_smote,y_smote)
# testing
y_pred=model.predict(x_test)
```

```python
print(classification_report(y_test,y_pred))
print('F1 Score',f1_score(y_test,y_pred))
```

#### 3. XGB Classifier (after class balance)

```python
model=XGBClassifier(n_jobs=1)

# training
model.fit(x_smote,y_smote)
# testing
y_pred=model.predict(x_test)

print(classification_report(y_test,y_pred))
print('F1 Score',f1_score(y_test,y_pred))
```

### Conclusion: so after balancing the data The Random Forest Classifier model Perform better compare to other two model with 82 % accuracy
